```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
import matplotlib
from matplotlib.pyplot import figure
from matplotlib.colors import ListedColormap
```

## Training Hours & Number of Parameters

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# figure size
figure(figsize=(7, 4), dpi=80)

# number of parameters 
x = [52.3,19,46.4,51.6,22.3,58.6,76.1,72.4] 
# training hours 
y = [26.4,15.2,24.4,25.6,18.3,29.2,35.8,36.5] 

# name of architectures 
classes = ['Our method', '3D UNet', '3D ResNet ', 'UNet ++', 'Attention', '3D ResUNet', '3DDeepLabV3', '3D PSP-Net']

# range of colosr values adjusted to the number of parameters 
values = [0, 1, 2, 3, 4, 5, 6, 7]
# color type
color = ListedColormap(['r','b','g'])

# increase the shape of parameter diameters in the figeure 
siz = []
for i in x:
    siz.append(i * 20)
sizes = siz

# assign general values to the plot
scatter = plt.scatter(x, y, s=sizes, c=values)

# title 
plt.title('Training Hours & Number of Parameters', size = 15)
# x label 
plt.xlabel('Parameters(million)')
# y label
plt.ylabel('Training hours')
# add legend to the lower right of the figure for the name of architectures 
plt.legend(loc='lower right', handles=scatter.legend_elements()[0], labels = classes)
# show color scale
plt.colorbar();  
```

## Accurecy & Number of Parameters

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# figure size
figure(figsize=(7, 4), dpi=80)

# number of parameters 
x = [52.3,19,46.4,51.6,22.3,58.6,76.1,72.4] 
# accurecy 
y = [0.8854, 0.8215, 0.8567, 0.8784, 0.8598, 0.8511, 0.8442, 0.8347] 

# name of architectures 
classes = ['Our method', '3D UNet', '3D ResNet ', 'UNet ++', 'Attention', '3D ResUNet', '3DDeepLabV3', '3D PSP-Net']

# range of colosr values adjusted to the number of parameters 
values = [0, 10, 20, 30, 40, 50, 60, 70]
# color type
color = ListedColormap(['r','g','b', 'g'])

# increase the shape of parameter diameters in the figeure 
siz = []
for i in x:
    siz.append(i * 20)
sizes = siz

# assign general values to the plot
scatter = plt.scatter(x, y, s=sizes, c=values)

# title 
plt.title('Dice Score & Number of Parameters', size = 15)
# x label 
plt.xlabel('Parameters(million)')
# y label
plt.ylabel('Dice Score')
# add legend to the lower right of the figure for the name of architectures 
plt.legend(loc='upper right', handles=scatter.legend_elements()[0], labels = classes)
# show color scale
plt.colorbar();  
```
